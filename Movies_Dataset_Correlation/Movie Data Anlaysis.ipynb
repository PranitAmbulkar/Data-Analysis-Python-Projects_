This is the movie data analysis project using Python

```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
plt.rcParams['figure.figsize']= (12,8) # for adjustment of the configuration of plots that we will create
```

```python
# Reading the data

movies = pd.read_csv("/content/drive/MyDrive/Data Science Analytics Project/Movie data analysis/movies.csv")
movies.head()
movies
```

**We will check for the missing data in the database**

```python
movies.isnull().sum()
```

```python
for col in movies.columns:
  pct_missing=round(np.mean(movies[col].isnull()),2)
  print('{} - {}%'.format(col, pct_missing))
```

```python
movies.isnull().sum()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(axis=0, inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.dtypes
```

**Changing the datatype of the columns**

```python
movies['budget']=movies['budget'].astype('int64')
```

```python
movies['gross']= movies['gross'].astype('int64')
```

```python
movies.head()
```

```python
# now we will fetch the year from the release date into the new column

movies['released'].astype('str')
```

```python
movies['year_correct'] = movies['released'].astype('str')
```

```python
movies.sample(3)
```

```python
movies['year_correct']=movies['year_correct'].str.extract(r'\b(\d{4})\b')[0]
```

```python
movies.sample(4)
```

```python
movies.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

**Checking for the Duplicate values**

```python
movies['company'] = movies['company'].drop_duplicates().sort_values(ascending=False)
```

```python
movies.drop_duplicates(inplace=True)
```

```python
movies['company'].value_counts()
```

```python
# Budget could be have high correlation with the gross
# Company also could have the high correlation with the gross


# Scatter plot with budget Vs Gross revenue


plt.scatter(x=movies['budget'], y=movies['gross'], color='green')
plt.title("Budget Vs Gross revenue")
plt.xlabel('Gross Earnings')
plt.ylabel("Budget for film")
plt.show()
```

```python
# Plot budget vs rows using seaborn

sns.regplot(x=movies['budget'], y=movies['gross'], data=movies, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.show()
```

```python
# Lets start looking at the correlation

movies.corr()
```

```python
# Pearson correlation
# kendell correlation
# spearman correlation

movies.corr(method='pearson')
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = movies.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.show()
```

```python
movies_numerized = movies

for col_name in movies_numerized.columns:
  if(movies_numerized[col_name].dtype == 'object'):
    movies_numerized[col_name] = movies_numerized[col_name].astype('category')
    movies_numerized[col_name] = movies_numerized[col_name].cat.codes
```

```python
movies_numerized.head()
```

```python
movies.head()
```

```python
correlation_matrix = movies_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Feature")

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = movies_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs


sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

- **Votes and bidget have the highest correlation**

- **Company has the lowest correlation**
